#similarity based nearest neighbor

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv',encoding='utf8')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data=df.copy()
```

```python
data.head()
```

```python
data.head(1)['summary']
```

```python
summ = data.dropna(axis = 0, subset = ['url'])
count = (summ.
     groupby(by = ['url'])['No_of_visit'].
     count().
     reset_index().
     rename(columns = {'No_of_visit': 'totalvisitCount'})
     [['url', 'totalvisitCount']]
    )
count.head()
```

```python
datawithcount = summ.merge(count, left_on = 'url', right_on = 'url', how = 'left')
datawithcount.head()
```

```python

pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(count['totalvisitCount'].describe())
```

```python

popularity_threshold = 2
rating_popular_website= datawithcount.query('totalvisitCount >= @popularity_threshold')
rating_popular_website.head()
```

```python
webiste_features_df=rating_popular_website.pivot_table(index='url',columns='userid',values='No_of_visit').fillna(0)
webiste_features_df.head()
```

```python
from scipy.sparse import csr_matrix

website_features_df_matrix = csr_matrix(webiste_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(website_features_df_matrix)
```

```python
webiste_features_df.shape
```

```python

query_index = np.random.choice(webiste_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(webiste_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors =2)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(webiste_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, webiste_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
