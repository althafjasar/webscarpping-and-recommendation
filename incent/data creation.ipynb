```python
def fun(leng,summary):
    data=[]
    
    for i in range(leng):
        data.append({
            'userid':np.random.randint(0,100),
            'Summary':np.random.choice(summary),
            'No_of_visit':np.random.randint(0,10)
            
        })
    
    return pd.DataFrame(data)   
            
```

```python
data=fun(1000,summary)
```
